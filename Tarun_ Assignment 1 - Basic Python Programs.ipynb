### 1.

Write a program to Find the Factorial of a given input Number

```python
number = int(input("Enter Number "))                          #User inputs for number you want to find Factorial
fact = 1                                                      #new variable taken as loop will multiply the number to it, 1 does not change value so taken 1

if number < 0:                                                 # if used as 0 does not have any factorial
  print("Fact of values below Zero, Does not Exist")
elif number == 0:                                              #Factorial of zero is always 1
  print("Factorial of 0 is always 1")
else:
  for i in range(1,number+1):                                  # for loop used to find
    fact = fact * i                                            # Multiply the value 
print("factorial is", fact)                                    # priting the final result
```

### 2. 

Write a program to check if the given year is a Leap Year or not

```python
# Leap year comes in every 4 years
# If we divide by 4 reminder should be 0,
# Using "If statement" for this and taking inputs from user:

x = int(input("Enter the year to check Lear year or Not : "))

if x % 4 == 0 and x % 100 != 0 :                            # not divided by 100 means century year (ending with 00) 
  print("Entered Year is : ", x)                            # century year divided by 400 is leap year
  print("Entered year is Leap Year")
elif x % 400 == 0 and x % 100 == 0:
  print("Entered Year is : ", x)                            # divided by 100 means century year (ending with 00) 
  print("Entered year is Leap Year")                        # divided by 4 is a leap year
else:
  print("Entered year is : ", x)                            # others are not a leap year those % is not zero
  print("Entered year is NOT a Leap Year")
```

Write a program to convert from Celsius To Fahrenheit

```python
# Using simple arithmetic operators to convert Celsius To Fahrenheit
# formula for converting Celsius To Fahrenheit is ('x' * 1.80 ) + 32
# taking inputs from the user

c = int(input("Enter Degree in Celsius : "))  # inputs from user
f = ((c * 1.80) + 32)                         # formula to convert

print(f, " Fahrenheit")                       # printing the statement
```

### 4.

Write a program to check if the given number is a palindrome number.

A palindrome number is a number that is same after reverse. For example, 121 is a palindrome number.

```python
# using if statement and List function to check reverse is same or not

p = input("Enter number to check : ")      # input from user, in the string format, as reverse of string or number, checking is same

x = p[::-1]                                # list function and reversing by using -1 to make it reverse

if p == x:                                 # using if and operators in the same and printing
  print("Answer : " , p , " is palindrome number")    
else:
  print("Answer : ",p ,"NOT a  palindrome number")
```
